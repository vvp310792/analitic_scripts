```r
#определение LD

library(httr)
library(jsonlite)
library(xml2)
library(dplyr)
library(stringr)
library(googlesheets4)
mygenetics_genes<-read_sheet("https://docs.google.com/spreadsheets/d/1SusCweQyufxlX84VS4DWIWsxDnUM0LxH2dbwmpDoTho/edit#gid=1723433368",sheet ="Genes")


all_snp_data<-NULL
for (i in 1:length(mygenetics_genes$data)){


server <- "https://rest.ensembl.org"
ext <- paste0("/ld/human/",mygenetics_genes$data[i],"/1000GENOMES:phase_3:CEU?")
 
r <- GET(paste(server, ext, sep = ""), content_type("application/json"))
 

if (str_detect(content(r)[[1]],"Could not fetch variation object for id:")==FALSE) 
{# use this if you get a simple nested list back, otherwise inspect its structure
# head(data.frame(t(sapply(content(r),c))))
snp_data<-fromJSON(toJSON(content(r)))
snp_data$r2<-as.numeric(snp_data$r2)
snp_data<-snp_data %>%
filter(r2>0.32)%>%
select(variation1,variation2,r2,d_prime,population_name)%>%
arrange(desc(r2))


snp_data_string<-snp_data[1,]
snp_data_string$variation2<-snp_data_string$variation1
snp_data_string$r2<-1
snp_data_string$d_prime<-"1.000000"
snp_data<-rbind(snp_data_string,snp_data)
snp_data
all_snp_data<-rbind(all_snp_data,snp_data)
} else {
  print(mygenetics_genes$data[i])
}
} 
```
